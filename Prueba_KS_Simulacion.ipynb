## Objetivo
Aplicar la prueba de Kolmogorov–Smirnov a un conjunto de números pseudoaleatorios para verificar si pueden considerarse provenientes de una distribución Uniforme(0,1).

```python
import numpy as np
import math
import matplotlib.pyplot as plt

# Datos (mismo conjunto de la tarea anterior)
data = [0.78961, 0.05230, 0.10699, 0.55877, 0.14151,
    0.76086, 0.12079, 0.27738, 0.65726, 0.79269,
    0.80548, 0.82654, 0.29453, 0.20852, 0.42989,
    0.58518, 0.98611, 0.34488, 0.34358, 0.11537,
    0.89898, 0.57880, 0.67621, 0.05010, 0.00121,
    0.28269, 0.73059, 0.70119, 0.18284, 0.49962,
    0.38618, 0.76910, 0.68334, 0.55170, 0.10850,
    0.79982, 0.45679, 0.21631, 0.87616, 0.55743,
    0.58962, 0.33216, 0.03185, 0.61168, 0.09264,
    0.69623, 0.17028, 0.05475, 0.91512, 0.76262,
    0.29931, 0.30861, 0.83358, 0.51781, 0.03272,
    0.57410, 0.26593, 0.85903, 0.43338, 0.35286,
    0.24000, 0.65559, 0.38507, 0.90829, 0.94187,
    0.93655, 0.88809, 0.81772, 0.36982, 0.19904,
    0.54325, 0.62400, 0.09133, 0.41678, 0.33954,
    0.58244, 0.85853, 0.88752, 0.33729, 0.15506,
    0.23949, 0.53559, 0.33381, 0.49383, 0.75103,
    0.19962, 0.65002, 0.74579, 0.79113, 0.63453,
    0.19147, 0.40644, 0.08128, 0.73435, 0.22724,
    0.22287, 0.07281, 0.64183, 0.44267, 0.72102
]

x = np.array(data, dtype=float)
n = len(x)

print("n =", n)
```

```python
import matplotlib
import matplotlib.pyplot as plt
print("matplotlib version:", matplotlib.__version__)
```

```python
import numpy as np
import math

# 정렬
xs = np.sort(x)
n = len(xs)

# D+ , D- 계산
D_plus = 0.0
D_minus = 0.0

for i in range(1, n+1):
    F_emp = i / n
    F_theo = xs[i-1]      # Uniform(0,1)에서 F(x)=x
    D_plus = max(D_plus, F_emp - F_theo)
    D_minus = max(D_minus, F_theo - (i-1)/n)

D = max(D_plus, D_minus)

# 임계값 (alpha=0.05)
alpha = 0.05
D_crit = 1.36 / math.sqrt(n)

print("n =", n)
print("D+ =", D_plus)
print("D- =", D_minus)
print("D  =", D)
print("D_crit =", D_crit)

if D <= D_crit:
    print("Conclusión: NO se rechaza H0. Los datos pueden provenir de Uniforme(0,1).")
else:
    print("Conclusión: Se rechaza H0. Los datos NO parecen provenir de Uniforme(0,1).")
```

```python
import matplotlib.pyplot as plt
import numpy as np

Fn = np.arange(1, n+1) / n

plt.step(xs, Fn, where="post", label="F_n(x) (Empírica)")
plt.plot(xs, xs, label="F(x)=x (Uniforme)")
plt.xlabel("x")
plt.ylabel("F(x)")
plt.title("Prueba Kolmogorov–Smirnov: EDF vs Teórica")
plt.legend()
plt.grid(True)
plt.show()
```

## Conclusión

Para un nivel de significancia α = 0.05 se obtuvo un estadístico
D = 0.05642 y un valor crítico D₍crit₎ = 0.136.

Dado que D ≤ D₍crit₎, **no se rechaza la hipótesis nula H₀**.
Por lo tanto, el conjunto de números analizados **puede considerarse
proveniente de una distribución Uniforme(0,1)**.
